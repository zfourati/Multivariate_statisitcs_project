```r
mqu_spe <- read.delim ('https://raw.githubusercontent.com/zdealveindy/anadat-r/master/data/MQU_spe.txt', row.names = 1)
mqu_env <- read.delim ('https://raw.githubusercontent.com/zdealveindy/anadat-r/master/data/MQU_env.txt', row.names = 1)
mqu_traits <- read.delim ('https://raw.githubusercontent.com/zdealveindy/anadat-r/master/data/MQU_traits.txt', row.names = 1)
mqu_loc <- read.delim ('https://raw.githubusercontent.com/zdealveindy/anadat-r/master/data/MQU_loc.txt')
 
```

```r
library(tidyverse)   
library(readr)       
library(readxl)     
library(vegan)       
library(reshape2)  
library(dplyr)
library(randomForest)
```

## Load data

```r
spe <- read_delim("MQU_spe.txt", delim = "\t")
env <- read_delim("MQU_env.txt", delim = "\t")
traits <- read_delim("MQU_traits.txt", delim = "\t")
loc <- loc <- read_delim("MQU_loc.txt", delim = "\t")
```

## Data exploration

```r
head(spe)  #importance value index (the higher number, the more important in the site)
head(env)
head(traits)
head (loc)
summary(traits)
summary(env)
```

```r
# dimension
dim(spe)
dim(env)
dim(traits)
dim(loc)
```

```r
# modify name of location by site
spe <- spe %>% 
  rename(site = `...1`)

env <- env %>%
  rename(site = `Locality code`)

loc <- loc %>%
  rename(site = `Locality code`)
```

```r
spe$site
env$site
loc$site
```

```r
# richness site
specnumber(spe[,-1])
```

```r
# cover by site
rowSums(spe[,-1])
```

## Visualization

```r
#richness per site
richness <- specnumber(spe[,-1])

barplot(richness, main="Species Richness by Site")
```

```r
# traits distribution
traits_long <- melt(traits, id.vars = "Species")

ggplot(traits_long, aes(value)) +
  geom_histogram(bins = 30) +
  facet_wrap(~variable, scales = "free") +
  theme_minimal()
```

```r
# boxplot traits
ggplot(traits_long, aes(x=variable, y=value)) +
  geom_boxplot(fill="lightblue") +
  facet_wrap(~variable, scales="free") +
  theme_minimal()
```

```r
# traits correlation
cor_mat <- cor(traits[,2:6])
corrplot::corrplot(cor_mat, method="color", type="upper")
```

```r
# environment vs site
ggplot(loc, aes(Longitude, Latitude)) +
  geom_point(size=3) +
  geom_text(aes(label=site), nudge_y=0.01) +
  theme_minimal()
```
